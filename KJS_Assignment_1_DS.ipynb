````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Read the data and display the first 5 rows

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
new=pd.get_dummies(['sex','embarked','pclass'])
new
titanic_data_new=pd.concat([titanic_data,new],axis=1)
titanic_data_new

````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
titanic_data_new.fillna(titanic_data_new.age.mean(),inplace=True)
titanic_data_new.fillna(titanic_data_new.fare.mean(),inplace=True)
titanic_data_new.info()
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
titanic_data_new['age_normalized']=(titanic_data_new['age']-titanic_data_new['age'].mean())/titanic_data_new['age'].std()
titanic_data_new['fare_normalized']=(titanic_data_new['fare']-titanic_data_new['fare'].mean())/titanic_data_new['fare'].std()
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
titanic_data_new.drop(['age','fare'],inplace=True)
titanic_data_new.info()
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````
plt.figure(figsize=(100,100))
y=titanic_data_new.age_normalized
x=titanic_data_new.index
plt.plot(x,y)
plt.show()
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
plt.figure(figsize=(10,10))
x=np.arange(-3*np.pi,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('X Values')
plt.ylabel('Sin(x)')
plt.title('Sine Wave')
plt.show()
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
x=np.arange(-2*np.pi,2*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
a=np.fft.fft(y)
b=np.fft.fft(z)
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,a)
plt.plot(x,b)
plt.xlabel('X values')
plt.ylabel('Sine Cosine and their Fourier Transforms')
plt.title('Sinusoidal Curves and their Fourier Transforms')
plt.show()
````
